```python
def collatz(n):
  count =0
  while n>1:
    if n%2 == 0: #Se for par
      n = n/2      
    else: #se Ímpar
      n = 3*n+1
    count+=1
  return(count)
```

```python
def collatz_check(n):
  passos = []
  passos.append(n)
  while n>1:
    if n%2 == 0: #Se for par
      n = n/2     
    else: #se Ímpar
      n = 3*n+1
    
    passos.append(n)
  return(passos)
```

```python
lst_collatz =[]

for n in range(2,10000000):
  lst_collatz.append(collatz(n))

#print(lst_collatz)
```
